1. COMPLEMENT OF IMAGE


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('image.jpg').astype(np.uint8)
```

```python
image=np.array([[100,200,100],
                [230,145,200],
                [105,155,175]])
```

```python
img_c=cv2.bitwise_not(image)
img_c1=cv2.bitwise_not(img)
```

```python
plt.subplot(2,2,1)
plt.imshow(image,cmap='gray')
plt.title('original image')
plt.subplot(2,2,2)
plt.imshow(img_c,cmap='gray')
plt.title('inverted image')
plt.subplot(2,2,3)
plt.imshow(img,cmap='gray')
plt.title('original image')
plt.subplot(2,2,4)
plt.imshow(img_c1,cmap='gray')
plt.title('inverted image')

plt.show()
```

2. LOG TRANSFORM WITH C=1

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image1 = np.array([[1, 10, 50],
    [100, 150, 200],
    [220, 240, 255]]).astype(np.float32)
```

```python
image1_transformed= 1 * np.log1p(image1)
```

```python
image1_norm=cv2.normalize(image1_transformed,None,0,255,cv2.NORM_MINMAX).astype(np.uint8)
```

```python
plt.subplot(1,2,1)
plt.imshow(image1,cmap='gray')
plt.title('original')

plt.subplot(1,2,2)
plt.imshow(image1_norm,cmap='gray')
plt.title('after log transform')

plt.show()
```

3. GAMMA TRANSFORM


```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img2=cv2.imread("image.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
img_float=img2/255.0
```

```python
gamma=int(input("enter gamma value"))
img_gamma=1* np.power(img_float,gamma)
```

```python
img2_transformed=(img_gamma*255).astype(np.uint8)
```

```python
plt.subplot(1,2,1)
plt.imshow(img2,cmap='gray',vmin=0, vmax=255)
plt.title('original')

plt.subplot(1,2,2)
plt.imshow(img2_transformed,cmap='gray',vmin=0, vmax=255)
plt.title('gamma transformed image')
```

4. HISTOGRAM EQUALISATION

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img3=cv2.imread('image.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
equalisedimg=cv2.equalizeHist(img3)
```

```python
hist_o=cv2.calcHist([img3], [0],None,[256],[0,256])
hist_i=cv2.calcHist([equalisedimg],[0],None,[256],[0,256])
```

```python
plt.figure(figsize=(12,6))
plt.subplot(2,2,1)
plt.imshow(img3,cmap='gray')
plt.title('original')
plt.axis('off')
plt.subplot(2,2,2)
plt.plot(hist_o)
plt.title('histogram original')
plt.xlabel('pixel value')
plt.ylabel('frequency')


plt.subplot(2,2,3)
plt.imshow(equalisedimg,cmap='gray')
plt.title('equalized')
plt.axis('off')
plt.subplot(2,2,4)
plt.plot(hist_i)
plt.title('histogram equalized')
plt.xlabel('pixel value')
plt.ylabel('frequency')
plt.show()
```

5. padding/border

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img4=cv2.imread('image.jpg',cv2.IMREAD_GRAYSCALE)
image4= np.array([[100, 150, 200],
                [120, 170, 220],
                [140, 190, 240]],dtype=np.uint8)
```

```python
border_img4=cv2.copyMakeBorder(img4,10,10,10,10,cv2.BORDER_CONSTANT,value=0)
bordered_image4 = cv2.copyMakeBorder(image4, 1, 3, 1, 2, cv2.BORDER_CONSTANT, value=220)
```

```python
plt.subplot(2,2,1)
plt.imshow(image4,cmap='gray')
plt.title('original')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(bordered_image4,cmap='gray')
plt.title('border')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(img4,cmap='gray')
plt.title('original')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(border_img4,cmap='gray')
plt.title('border')
plt.axis('off')

plt.show()
```

6. rgb to grayscale


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img5=cv2.imread('image2.jpeg')
```

```python
gray_mean1=np.mean(img5,axis=2)
```

```python
wr=float(input("enter r"))
wg=float(input("enter g"))
wb=float(input("enter b"))

gray_weighted1 = (wr * img5[:, :, 2] + wg * img5[:, :, 1] + wb * img5[:, :, 0]).astype(np.uint8)
```

```python
plt.subplot(1,3,1)
plt.imshow(img5)
plt.title('original')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(gray_mean1,cmap='gray')
plt.title('mean')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(gray_weighted1,cmap='gray')
plt.title('weighted')
plt.axis('off')

plt.show()
```

7. grayscale to rgb

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img6=cv2.imread('image.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
mean_threshold=img6.mean()
ret,binary_img=cv2.threshold(img6,mean_threshold,255,cv2.THRESH_BINARY)

user_threshold=int(input("enter threshold value"))
ret,binary_img1=cv2.threshold(img6,user_threshold,255,cv2.THRESH_BINARY)
```

```python
plt.subplot(1,3,1)
plt.imshow(img6)
plt.title('original')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(binary_img,cmap='gray')
plt.title('mean')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(binary_img1,cmap='gray')
plt.title('weighted')
plt.axis('off')

plt.show()
```

8. intensity level slicing

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img7=cv2.imread('image.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
rmin=int(input("enter value"))
rmax=int(input("enter value"))
smin=int(input("enter value"))
smax=int(input("enter value"))
```

```python
streched_img=smin+(smax-smin)*((img7-rmin)/(rmax-rmin))
```

```python
img7_f=np.clip(streched_img,0,255).astype(np.uint8)
```

```python
plt.subplot(1,2,1)
plt.imshow(img7,cmap='gray')
plt.title('original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img7_f,cmap='gray')
plt.title('streched')
plt.axis('off')
```
